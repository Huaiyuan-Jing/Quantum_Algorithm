 (2):

 #### Step 1: Initial State
 $$
|\psi\rangle = \cos\frac{\theta}{2}|0\rangle + e^{i\phi}\sin\frac{\theta}{2}|1\rangle
$$

---

#### Step 2: Apply $X$ Gate
The $X$ gate (Pauli-X) swaps $|0\rangle$ and $|1\rangle$:
$$
X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}
$$

After applying $X$:
$$
|\psi'\rangle = X|\psi\rangle = \cos\frac{\theta}{2}|1\rangle + e^{i\phi}\sin\frac{\theta}{2}|0\rangle
$$
Rewriting:
$$
|\psi'\rangle = e^{i\phi}\sin\frac{\theta}{2}|0\rangle + \cos\frac{\theta}{2}|1\rangle
$$

---

#### Step 3: Apply $T$ Gate
The $T$ gate is a phase gate defined as:
$$
T = \begin{bmatrix} 1 & 0 \\ 0 & e^{i\pi/4} \end{bmatrix}
$$

Applying $T$:
$$
|\psi''\rangle = T|\psi'\rangle = e^{i\phi}\sin\frac{\theta}{2}|0\rangle + \cos\frac{\theta}{2}e^{i\pi/4}|1\rangle
$$

---

#### Step 4: Apply $H$ Gate
The $H$ gate (Hadamard) is defined as:
$$
H = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
$$

Applying $H$ to $|\psi''\rangle$:
$$
|\psi'''\rangle = H|\psi''\rangle
$$

Substitute the state:
$$
|\psi'''\rangle = H\left(e^{i\phi}\sin\frac{\theta}{2}|0\rangle + \cos\frac{\theta}{2}e^{i\pi/4}|1\rangle\right)
$$

Use $H|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ and $H|1\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$:
$$
|\psi'''\rangle = \frac{1}{\sqrt{2}} \left(e^{i\phi}\sin\frac{\theta}{2}(|0\rangle + |1\rangle) + \cos\frac{\theta}{2}e^{i\pi/4}(|0\rangle - |1\rangle)\right)
$$

Combine terms:
$$
|\psi'''\rangle = \frac{1}{\sqrt{2}} \left[\left(e^{i\phi}\sin\frac{\theta}{2} + \cos\frac{\theta}{2}e^{i\pi/4}\right)|0\rangle + \left(e^{i\phi}\sin\frac{\theta}{2} - \cos\frac{\theta}{2}e^{i\pi/4}\right)|1\rangle\right]
$$
