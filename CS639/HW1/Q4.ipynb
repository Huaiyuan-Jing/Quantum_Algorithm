### Part A:

Alice begins with a quantum state $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$ and shares a maximally entangled pair, $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$, with Bob.

The combined initial state of the system, which includes Alice’s qubit $|\psi\rangle$ and the shared entangled pair $|\Phi^+\rangle$ represent the full system of three qubits. Measurement of Alice’s two qubits projects the system into one of four possible states. Derive the resulting state of Bob’s qubit for each measurement outcome.

Based on Alice’s measurement outcome, Bob must apply the unitary operation $U$ (from the set $\{I, X, Z, XZ\}$) to recover the original state $|\psi\rangle$. After Bob applies this correction, calculate the fidelity of the teleported state using the formula, $F = |\langle \psi | \phi \rangle|^2,$ where $|\phi\rangle$ is Bob’s qubit state after the correction for each of the possible scenarios of Alice's measurement.

### Part A Solution: (TODO)

The fidelity of the teleported state is given by:
$$
F = |\langle \psi | \phi \rangle|^2
$$
Since Bob’s final state is always restored to $|\psi\rangle$, we have:
$$
F = |\langle \psi | \psi \rangle|^2 = 1
$$


### Part B:

### Part B Solution: (TODO)

Alice and Bob share the entangled state
$$
|\Phi(\theta)\rangle_{23} = \cos\theta\,|00\rangle_{23} + \sin\theta\,|11\rangle_{23},
$$
with $0<\theta\le \pi/4$. (When $\theta=\pi/4$ the state is maximally entangled.)  
The state to be teleported is an arbitrary qubit
$$
|\psi\rangle_1 = \alpha|0\rangle_1 + \beta|1\rangle_1,
$$
with $|\alpha|^2+|\beta|^2=1$.

Thus the total three-qubit state is
$$
\begin{aligned}
|\Psi\rangle_{123} &= |\psi\rangle_1\otimes |\Phi(\theta)\rangle_{23} &= \Bigl(\alpha|0\rangle_1 + \beta|1\rangle_1\Bigr) \otimes \Bigl(\cos\theta\,|00\rangle_{23}+\sin\theta\,|11\rangle_{23}\Bigr) &=\alpha\cos\theta\,|0,0,0\rangle + \alpha\sin\theta\,|0,1,1\rangle +\beta\cos\theta\,|1,0,0\rangle + \beta\sin\theta\,|1,1,1\rangle.
\end{aligned}
$$
Here qubits 1 and 2 are held by Alice and qubit 3 by Bob.

---

### **Alice’s Measurement in the Computational Basis**

#### **Outcome 00**

The outcome $00$ means that Alice’s qubits are found in the state $|0\rangle_1|0\rangle_2$. Looking at the expanded state,
- The only term that has qubits 1 and 2 in $|0,0\rangle$ is
  $$
  \alpha\cos\theta\,|0,0,0\rangle.
  $$
  
Thus, upon outcome $00$ the (unnormalized) post‐measurement state is
$$
|\phi_B^{00}\rangle = \alpha\cos\theta\,|0\rangle_3.
$$
The probability for this outcome is
$$
P(00)=|\alpha\cos\theta|^2.
$$
After normalizing (provided $P(00)\neq0$), Bob’s state becomes
$$
|\tilde\phi_B^{00}\rangle = |0\rangle_3.
$$

Now, the fidelity of teleportation (i.e. the overlap of Bob’s state with the original state $|\psi\rangle$) is given by
$$
F_{00} = |\langle\psi|0\rangle|^2.
$$
Since
$$
\langle \psi|0\rangle = \alpha^*,
$$
we obtain
$$
F_{00}=|\alpha|^2=\frac{1}{2}
$$

---

#### **Outcome 01**

The outcome $01$ corresponds to Alice’s qubits being in state $|0\rangle_1|1\rangle_2$. From the total state, the only term with qubits 1 and 2 in $|0,1\rangle$ is
$$
\alpha\sin\theta\,|0,1,1\rangle.
$$
Thus, Bob’s (unnormalized) state is
$$
|\phi_B^{01}\rangle = \alpha\sin\theta\,|1\rangle_3.
$$
The probability for outcome $01$ is
$$
P(01)=|\alpha\sin\theta|^2.
$$
Normalizing, Bob’s state is
$$
|\tilde\phi_B^{01}\rangle = |1\rangle_3.
$$
Its fidelity with respect to $|\psi\rangle$ is
$$
F_{01} = |\langle \psi |1\rangle|^2.
$$
Since
$$
\langle \psi|1\rangle = \beta^*,
$$
we have
$$
F_{01}=|\beta|^2=\frac{1}{2}
$$
