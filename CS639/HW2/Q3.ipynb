## **Part B:**

### **Step 1: Depolarizing Noise Model**
Depolarizing noise on a single qubit applies a Pauli error ($X, Y, Z$) with probability $p/3$ each, and with probability $1 - p$, the qubit remains unchanged. This process can be described as:
$$
\mathcal{E}(\rho) = (1 - p) \rho + \frac{p}{3} (X \rho X + Y \rho Y + Z \rho Z)
$$
where the Pauli matrices are:
$$
X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}, \quad
Y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix}, \quad
Z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}.
$$

We will apply depolarizing noise after each gate in the circuit.

---

### **Step 2: Apply Depolarizing Noise After the Hadamard Gate**
From Part A, the state after the Hadamard gate is:
$$
\rho_1 = \frac{1}{2} \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}
$$

Applying depolarizing noise:
$$
\rho_1' = (1 - p_H) \rho_1 + \frac{p_H}{3} (X \rho_1 X + Y \rho_1 Y + Z \rho_1 Z).
$$
Expanding each term:
- $X \rho_1 X = \frac{1}{2} \begin{bmatrix} 1 & -1 \\ -1 & 1 \end{bmatrix}$
- $Y \rho_1 Y = \frac{1}{2} \begin{bmatrix} 1 & -1 \\ -1 & 1 \end{bmatrix}$
- $Z \rho_1 Z = \frac{1}{2} \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} = \rho_1$

Averaging these:
$$
X \rho_1 X + Y \rho_1 Y + Z \rho_1 Z = \frac{1}{2} \begin{bmatrix} 3 & -1 \\ -1 & 3 \end{bmatrix}.
$$
Thus, the noisy state after Hadamard is:
$$
\rho_1' = (1 - p_H) \rho_1 + \frac{p_H}{6} \begin{bmatrix} 3 & -1 \\ -1 & 3 \end{bmatrix}.
$$

---

### **Step 3: Apply the $R_y(\theta)$ Gate**
The $R_y(\theta)$ gate transforms $\rho_1'$ into:
$$
\rho_2' = R_y(\theta) \rho_1' R_y^\dagger(\theta).
$$
Since depolarizing noise is basis-independent, we apply it again:
$$
\rho_2'' = (1 - p_R) \rho_2' + \frac{p_R}{3} (X \rho_2' X + Y \rho_2' Y + Z \rho_2' Z).
$$

After simplification (similar to Step 2), the final state takes the form:
$$
\rho_{\text{final}} = (1 - p_H)(1 - p_R) \rho_2 + (1 - (1 - p_H)(1 - p_R)) \frac{I}{2}.
$$

This shows that the final noisy state is a probabilistic mixture of the noiseless final state $\rho_2$ and the maximally mixed state $\frac{I}{2}$.

---

### **Conclusion**
Under depolarizing noise, the final state can be written as:
$$
\rho_{\text{final}} = (1 - p_{\text{eff}}) \rho_2 + p_{\text{eff}} \frac{I}{2},
$$
where
$$
p_{\text{eff}} = 1 - (1 - p_H)(1 - p_R).
$$
This confirms that the effect of depolarizing noise is to mix the ideal state with the maximally mixed state, as expected.

# **Part C**:
### **Step 1: Compute Purity Before Introducing Depolarizing Noise**
Purity is defined as:
$$
\text{Purity} = \text{Tr}(\rho^2).
$$
From Part A, the final density matrix **before noise** is $\rho_2$:
$$
\rho_2 = R_y(\theta) \rho_1 R_y^\dagger(\theta).
$$
Since $\rho_2$ represents a pure state, it satisfies:
$$
\text{Tr}(\rho_2^2) = 1.
$$
Thus, before noise, the purity is always **1**, confirming that the state is pure.

---

### **Step 2: Compute Purity After Depolarizing Noise**
After introducing depolarizing noise, the final density matrix becomes:
$$
\rho_{\text{final}} = (1 - p_{\text{eff}}) \rho_2 + p_{\text{eff}} \frac{I}{2},
$$
where $p_{\text{eff}} = 1 - (1 - p_H)(1 - p_R)$.

Now, computing $\text{Tr}(\rho_{\text{final}}^2)$:
$$
\text{Tr} \left( \left( (1 - p_{\text{eff}}) \rho_2 + p_{\text{eff}} \frac{I}{2} \right)^2 \right).
$$
Expanding the square:
$$
\text{Tr} \left( (1 - p_{\text{eff}})^2 \rho_2^2 + 2(1 - p_{\text{eff}}) p_{\text{eff}} \rho_2 \frac{I}{2} + p_{\text{eff}}^2 \frac{I}{4} \right).
$$
Since $\text{Tr}(\rho_2^2) = 1$ and $\text{Tr}(\rho_2 I) = \text{Tr}(\rho_2) = 1$, and $\text{Tr}(I^2) = 2$, we get:
$$
(1 - p_{\text{eff}})^2 + 2(1 - p_{\text{eff}}) p_{\text{eff}} \frac{1}{2} + p_{\text{eff}}^2 \frac{1}{2}.
$$
Simplifying:
$$
\text{Purity} = (1 - p_{\text{eff}})^2 + \frac{p_{\text{eff}}}{2}.
$$
This shows that **as $p_{\text{eff}}$ increases, purity decreases**, indicating a transition from a pure quantum state to a classical mixed state.

---

### **Step 3: Minimum Value of Purity**
The minimum purity occurs when **maximal depolarizing noise** is applied, i.e., $p_{\text{eff}} = 1$. In that case:
$$
\text{Purity} = \left(1 - 1\right)^2 + \frac{1}{2} = \frac{1}{2}.
$$
Thus, the lowest possible purity is **$1/2$**, corresponding to a maximally mixed state:
$$
\rho_{\text{final}} = \frac{I}{2}.
$$
This represents complete classical randomness, meaning the quantum system has fully decohered into a classical mixture.

---

### **Conclusion**
- **Before noise:** $\text{Purity} = 1$ (state is pure).
- **After noise:** $\text{Purity} = (1 - p_{\text{eff}})^2 + p_{\text{eff}}/2$ (purity decreases with noise).
- **Minimum purity:** $\frac{1}{2}$, occurring when $p_{\text{eff}} = 1$, representing a fully mixed state.

This confirms that increasing depolarizing noise drives the system toward a classical mixed state, destroying quantum coherence.

## **Part D:**
### **Step 1: Compute Probability of Measuring $|0\rangle$ Before Noise**
In quantum mechanics, the probability of measuring a quantum state $\rho$ in the $|0\rangle$ basis is given by:

$$
P(0) = \text{Tr}(M_0 \rho)
$$

where $ M_0 = |0\rangle \langle 0| = \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} $.

From Part A, the final density matrix **before noise** is:

$$
\rho_2 = R_y(\theta) \rho_1 R_y^\dagger(\theta).
$$

Since the initial state is $ \rho_1 = |+\rangle \langle +| $, applying the $ R_y(\theta) $ gate gives:

$$
\rho_2 = \begin{bmatrix} \cos^2(\theta/2) & \cos(\theta/2) \sin(\theta/2) \\ \cos(\theta/2) \sin(\theta/2) & \sin^2(\theta/2) \end{bmatrix}.
$$

Computing $ P(0) $:

$$
P(0) = \text{Tr} \left( \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} \cos^2(\theta/2) & \cos(\theta/2) \sin(\theta/2) \\ \cos(\theta/2) \sin(\theta/2) & \sin^2(\theta/2) \end{bmatrix} \right).
$$

Since matrix trace sums only the diagonal elements:

$$
P(0) = \cos^2(\theta/2).
$$

---

### **Step 2: Compute Probability of Measuring $|0\rangle$ After Depolarizing Noise**
After introducing depolarizing noise, the final density matrix is:

$$
\rho_{\text{final}} = (1 - p_{\text{eff}}) \rho_2 + p_{\text{eff}} \frac{I}{2}.
$$

Using $ M_0 $, the probability of measuring $|0\rangle$ is:

$$
P'(0) = \text{Tr} \left( M_0 \rho_{\text{final}} \right).
$$

Expanding:

$$
P'(0) = (1 - p_{\text{eff}}) \cos^2(\theta/2) + p_{\text{eff}} \cdot \frac{1}{2}.
$$

---

### **Step 3: Comparing the Effects of Noise**
The noise affects the measurement probability in the following ways:

1. **Without noise $(p_{\text{eff}} = 0$)**:
   $$
   P(0) = \cos^2(\theta/2).
   $$
   This represents the **ideal measurement probability**.

2. **With noise ($p_{\text{eff}} > 0$)**:
   $$
   P'(0) = (1 - p_{\text{eff}}) \cos^2(\theta/2) + p_{\text{eff}} \cdot \frac{1}{2}.
   $$
   The probability now it not accurately $\cos^2(\theta/2)$.

3. **At maximum noise $(p_{\text{eff}} = 1$)**:
   $$
   P'(0) = \frac{1}{2}.
   $$
   This means the measurement is completely random, and the quantum state has fully decohered into a classical mixture.

---

### **Conclusion: How Noise Affects Measurement Reliability**
- **Noise shifts the probability toward 50%, reducing confidence in measurement outcomes.**
- **For small $ p_{\text{eff}} $, measurements are still somewhat reliable, but as noise increases, the outcome becomes increasingly random.**
- **At $ p_{\text{eff}} = 1 $, all measurement results are completely random, meaning the system has lost all quantum coherence.**

This demonstrates why **error mitigation and quantum error correction** are crucial for preserving reliable quantum computations.

## **Part E:**
- With probability $p_s$, the qubit is correctly prepared in the state $|0\rangle$.
- With probability $1-p_s$, the qubit is mistakenly prepared in the state $|1\rangle$.

Thus, the density matrix for a single qubit is given by:
$$
\rho = p_s |0\rangle\langle 0| + (1-p_s)|1\rangle\langle 1| = \begin{bmatrix} p_s & 0 \\ 0 & 1-p_s \end{bmatrix}.
$$

Since the state preparation noise is independent for each qubit, the two-qubit density matrix is the tensor product of the individual density matrices:
$$
\rho_{2q} = \rho \otimes \rho = \begin{bmatrix} p_s & 0 \\ 0 & 1-p_s \end{bmatrix} \otimes \begin{bmatrix} p_s & 0 \\ 0 & 1-p_s \end{bmatrix}.
$$

Evaluating the tensor product, we obtain:
$$
\rho_{2q} = \begin{bmatrix}
p_s^2 & 0 & 0 & 0 \\
0 & p_s(1-p_s) & 0 & 0 \\
0 & 0 & p_s(1-p_s) & 0 \\
0 & 0 & 0 & (1-p_s)^2
\end{bmatrix}.
$$

This density matrix represents a classical mixture of the computational basis states:
- The probability of state $|00\rangle$ is $p_s^2$.
- The probability of state $|01\rangle$ is $p_s(1-p_s)$.
- The probability of state $|10\rangle$ is $p_s(1-p_s)$.
- The probability of state $|11\rangle$ is $(1-p_s)^2$.
