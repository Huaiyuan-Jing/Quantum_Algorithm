```python
# Add your simulation code here.
# Measure each qubit

# Set up the simulator backend
simulator = AerSimulator()
# Execute the circuit on the simulator
job = simulator.run(circ, backend=simulator, shots=1024)
result = job.result()

# Get the counts (measurement outcomes)
counts = result.get_counts(circ)

# Plot the histogram of results
plot_histogram(counts)
```

```python
# Qiskit Aer imports

from qiskit_aer.noise import (NoiseModel, QuantumError, ReadoutError, pauli_error, depolarizing_error, thermal_relaxation_error)
```

```python
# Create a new NoiseModel
noise_model = NoiseModel()

# --- 1. Define the bit-flip quantum errors ---

# Single-qubit gate bit-flip error: p = 0.001 (0.1%)
# Pauli error: apply 'X' with probability 0.001, do nothing 'I' with probability 0.999
bit_flip_error_single = pauli_error([
    ('X', 0.001),  # flip
    ('I', 0.999)   # no flip
])

# Two-qubit gate bit-flip error: p = 0.02 (2%)
bit_flip_error_two = pauli_error([
    ('XX', 0.02),  # flip
    ('II', 0.98)   # no flip
])

# --- 2. Define the readout (measurement) error ---

# Readout error: 5% chance to flip the measured outcome
# This is a 2x2 matrix for 1 qubit:
#   P(measure 0 | true 0) = 0.95
#   P(measure 1 | true 0) = 0.05
#   P(measure 0 | true 1) = 0.05
#   P(measure 1 | true 1) = 0.95
readout_error = ReadoutError([
    [1 - 0.05, 0.05],
    [0.05,     1 - 0.05]
])

# --- 3. Add errors to the noise model ---

# Add the single-qubit bit-flip error to all single-qubit gate names of interest
# (Qiskit names vary depending on version; adjust as needed)
single_qubit_gates = ['x', 'y', 'z', 'h', 's', 'sdg', 't', 'tdg',
                      'rx', 'ry', 'rz', 'sx', 'sxdg', ]
for gate in single_qubit_gates:
    noise_model.add_all_qubit_quantum_error(bit_flip_error_single, gate)

# Add the two-qubit bit-flip error to all two-qubit gates (commonly 'cx')
two_qubit_gates = ['cx', 'cz', 'swap', 'rzz']
for gate in two_qubit_gates:
    noise_model.add_all_qubit_quantum_error(bit_flip_error_two, gate)
# Add readout error for all qubits (if you have N qubits, you can list them all or do it in a loop)
# Example for 3 qubits:
noise_model.add_all_qubit_readout_error(readout_error)

# Now 'noise_model' can be passed to a QasmSimulator in Qiskit Aer
# For example:
#   from qiskit_aer import AerSimulator
#   simulator = AerSimulator(noise_model=noise_model)
#
# Then you can execute circuits on 'simulator' to simulate these errors.
print(noise_model)
```

```python
# Perform noisy simulations here
simulator = AerSimulator(noise_model=noise_model)
circ5 = circ.copy()
circ5.measure_all()
circ5.save_statevector()
job = simulator.run(circ5, backend=simulator, shots=100000)
result = job.result()
counts5 = result.get_counts(circ5)
plot_histogram(counts5)
```

```python
import time
circ6 = circ.copy()
circ6.measure_all()
circ6.save_statevector()
# Perform the runtime comparison here
simulator = AerSimulator()
runtime_noiseless = time.time()
job_noiseless = simulator.run(circ6, shots=100000)
result_noiseless = job_noiseless.result()
runtime_noiseless = time.time() - runtime_noiseless
print("Runtime of noiseless simulation: ", runtime_noiseless)
simulator = AerSimulator(noise_model=noise_model)
runtime_noisy = time.time()
job_noisy = simulator.run(circ6, shots=100000)
result_noisy = job_noisy.result()
runtime_noisy = time.time() - runtime_noisy
print("Runtime of noisy simulation: ", runtime_noisy)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
# Plot each histogram on a separate axis
plot_histogram(result_noiseless.get_counts(), ax=axs[0])
axs[0].set_title("Circuit 1")

plot_histogram(result_noisy.get_counts(), ax=axs[1])
axs[1].set_title("Circuit 2")

# Show the combined figure
plt.show()
```
